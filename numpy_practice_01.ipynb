#### Copying Arrays

```python
# by default, arrays are copied by reference
a = [1,2,3]
b = a
b[0] = 100
a[0]
```

```python
#To fix it, use copy()
a[0] = 1
b = a.copy()
b[0] = 1000
a[0]
```

###### Boolean Masking and Advanced Indexing

```python
a = np.array([[13,45,34,56, 67, 13], [23, 69, 100, 120, 45, 56]])
a
```

```python
a > 50 # This outputs a list of Booleans
```

```python
(a > 50) & (a < 100) # This outputs a list of Booleans
```

```python
~((a > 50) & (a < 100)) # This outputs a list of Booleans
```

```python
a[(a > 50) & (a < 100)] #Here we are indexing with a List of Booleans. 
```

```python
np.any(a > 50, axis= 0)
```

```python
np.any(a > 100, axis= 1)
```
